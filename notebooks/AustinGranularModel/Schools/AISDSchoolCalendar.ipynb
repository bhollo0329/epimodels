```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
aisd_2018_2021_cal = pd.read_csv('/Users/kpierce/epimodels/notebooks/AustinGranularModel/Schools/data/AISDCalendars/AISD_2018_2021_Calendar.csv')
```

```python
aisd_2018_2021_cal.head()
```

```python
min_date = datetime(year=2018, month=1, day=1)
max_date = datetime(year=2021, month=12, day=31)
```

```python
date_range = (max_date - min_date)
date_range.days
```

```python
date_list = [max_date - timedelta(days=x) for x in range(date_range.days+1)]
weekday = {i: (True if i.weekday() < 5 else False) for i in date_list}
weekday_df = pd.DataFrame.from_dict([weekday]).transpose().reset_index()
weekday_df.columns = ['Date', 'is_weekday']
```

```python
weekday_df['Date'] = weekday_df['Date'].astype('datetime64[ns]')
aisd_2018_2021_cal['Date'] = aisd_2018_2021_cal['Date'].astype('datetime64[ns]')
```

```python
full_calendar = pd.merge(
    weekday_df,
    aisd_2018_2021_cal,
    on='Date',
    how='left'
)
```

```python
full_calendar['school_closed'] = [
    True if (row['is_weekday'] == False) or (row['StudentHoliday'] == True) else False for i, row in full_calendar.iterrows()
]
```

```python
full_calendar.sort_values('Date').head(40)
```

```python
full_calendar = full_calendar.rename(columns={'Date': 'date'})
```

```python
full_calendar.to_csv('/Users/kpierce/epimodels/notebooks/AustinGranularModel/Schools/data/AISDCalendars/AISD_2018_2021_Calendar_all_dates.csv')
```
